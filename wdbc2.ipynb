```r
wdbc=read.table("wdbc.txt", header=T,sep=",")
attach(wdbc)
for (i in 3:32)
{
cat(paste("c",i-2,sep=""),"total mean=g", round(mean(wdbc[,i]),4),"\n")
cat("B mean=", round(mean(wdbc[diagnosis=="B",i]),4),"\n")
cat("M mean=", round(mean(wdbc[diagnosis=="M",i]),4),"\n")
}
```

```r
wdbc=read.table("wdbc.txt", header=T,sep=",")
attach(wdbc)
pdf("output.pdf")
par(mai=c(0.5,0.5,0.5,0.5),mfcol=c(2,3))  
for (i in 3:32)
{
boxplot(wdbc[,i]~diagnosis,main=paste("c",i-2,sep=""),col=c("yellowgreen","gray"))
boxplot(wdbc[,i], main=paste("c",i-2,sep=""), col="yellow",boxwex=0.3)
boxplot(wdbc[diagnosis=="M",i],col="blue",boxwex= 0.1,	at= c(1.3), add=TRUE, axes=F)
boxplot(wdbc[diagnosis=="B",i],col="pink",boxwex= 0.1,	at= c(0.7), add=TRUE, axes=F)
}
graphics.off()
```
