```r
#複雜
babies=read.table("babies.txt", header=T)
summary(babies)
label=colnames(babies)
for (i in c(1,2,4:6))
{
cat(label[i],"mean=", mean(babies[ , 1],na.rm=T),
       "std=", sd(babies[ ,1] ,na.rm=T),"\n")
cat("smoke:",mean(babies[babies$smoke==1 &!is.na(babies$smoke), i],na.rm=T),"nosmoke:"
,mean(babies[babies$smoke==0 &!is.na(babies$smoke), i],na.rm=T),"\n")
}
```

```r
#練習
mean(babies[babies$smoke==1 &!is.na(babies$smoke), 1])
mean(babies[babies$smoke==0 &!is.na(babies$smoke), 1])
babies[babies$age<18 & !is.na(babies$age) , ]
```

```r
#精簡
babies=read.table("babies.txt", header=T)

babies$smoke=as.factor(babies$smoke)
summary(babies)
nrow(babies)
table(babies$smoke)
babies=na.exclude(babies)
label=colnames(babies)
for (i in c(1,2,4:6))
{
cat(label[i],"mean=", mean(babies[ , 1],na.rm=T),
       "std=", sd(babies[ ,1]),"\n")
cat("smoke:",mean(babies[babies$smoke==1, i]),"nosmoke:"
,mean(babies[babies$smoke==0, i]),"\n")
}
```
