```r
attach(iris)
par(mfrow=c(2,2))
label=colnames(iris)
for(i in 1:4)
{
hist(iris[,i],main=label[i])
st=shapiro.test(iris[,i])
cat(label[i],st$p.value,ifelse(st$p.value<0.05,"not norm","norm"),"\n")
}

st=shapiro.test(Sepal.Length)
names(st)
st$p.value

shapiro.test(Sepal.Width)
shapiro.test(Petal.Length)
shapiro.test(Petal.Width)
```
