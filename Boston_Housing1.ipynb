```r
boston = read.table("Boston_Housing.csv", sep = ",", header = T)
summary(boston)
attach(boston)
boston $ CLASS = ifelse ( MEDV > 30, "RICH", "POOR" )
boston $ CLASS = factor (boston $ CLASS)
summary(boston)
attach(boston)

chisq.test(table(CHAS, CLASS))
cor.test(CRIM, MEDV)

var.test(CRIM~CLASS)
t.test(CRIM~CLASS ,var.equal=F)
```

```r
#y=-0.3741xcomputerR-1.653xoldR ......
boston = read.table("Boston_Housing.csv", sep = ",", header = T)
model0 = lm(MEDV~ . , data=boston)
summary(step(model0))
```

```r
# rebuild model
model0 = lm(MEDV~ CRIM + ZN + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT, data=boston)
summary(step(model0))
```

```r
#predict
bostonnew = read.csv("Boston_Housing.csv", header = T)
pred=predict(model0,newdata=bostonnew)
all=data.frame(bostonnew,pred)
all
```
