```r
boston=read.csv("Boston_Housing.csv",header=T)
summary(boston)
install.packages("NbClust")
library(NbClust)
result=NbClust(boston,distance="euclidean",min.nc=2,max.nc=6,method="kmeans", index="all")
result$Best.partition
bostonall=data.frame(boston,Clust=result$Best.partition)
bostonall

label=colnames(bostonall)
for (i in 1:14)
{
f=var.test(bostonall[,i]~bostonall[,15]) # Clust
p=f$p.value
if (p<0.05)
	{	t=t.test(bostonall[,i]~bostonall[,15],var.equal=F)	}
else 
	{	t=t.test(bostonall[,i]~bostonall[,15],var.equal=T)	}
cat(label[i],round(t$p.value,4), ifelse(t$p.value<0.05,"*"," "),"\n")
}

names(t)
t$estimate
```
