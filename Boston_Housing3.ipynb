```r
boston=read.csv("Boston_Housing.csv",header=T)
summary(boston)
#regression
model=lm(MEDV ~ . ,data=boston)
summary(step(model))
#decision tree CART
install.packages("rpart")
library(rpart)
model1=rpart(MEDV ~ . ,data=boston)
model1
plot(model1)
text(model1)
# logistic regression
boston$LEVEL=ifelse(boston$MEDV>35, 1, 0)
boston=boston[ , -14]
```

```r
train.result=glm(LEVEL~.,data=boston,family=binomial(link=logit))
summary(train.result)
exp(train.result$coef) #exp(參數估計值)計算 odds ratio
```

```r
# C5.0 decision tree
boston$LEVEL=factor(boston$LEVEL)
install.packages("C50")
library(C50)
tree = C5.0(LEVEL~ . , data=boston)
summary(tree)
plot(tree)
```

```r
# association rule
attach(boston)
index=1+(RM>mean(RM))
boston$RM2=c("less","more")[index]
index2=1+(DIS>mean(DIS))
DIS2=c("short","long")[index2]
index3=1+(LSTAT>mean(LSTAT))
LSTAT2=c("poor","rich")[index2]
head(boston)
boston4 = boston[  ,8:14]
head(boston4)
summary(boston4)
install.packages("arules")
library(arules)
rule=apriori(boston4)
inspect(rule)
```

```r
rule=apriori(boston4,parameter=list(minlen=2,supp=0.1,conf=0.8),
appearance=list(rhs=c("LEVEL=1","LEVEL=0"),default="lhs"))
rulesort=sort(rule,by="lift")
inspect(rulesort)
```
