```r
babies=read.table("babies.txt", header=T)
summary(babies)
attach(babies)
babies$agegroup=1+(age>20)+(age>=25)+(age>=30)+(age>=35)+(age>=40)
attach(babies)
babies$agegroup=as.factor(agegroup)
summary(babies)
```

```r
label=c("-19","20-24","25-29","30-34","35-39","40-")
babies$group=label[agegroup]
head(babies,50)
babies$group=as.factor(babies$group)
summary(babies)
table(babies$group)
```

```r
babies1=subset(babies,agegroup==1)

babies1=babies[ babies$agegroup==1 & !is.na(babies$agegroup), ]
nrow(babies1)
write.table(babies1,"babies1.txt")
```

```r
babies2=babies[ babies$agegroup==2 & !is.na(babies$agegroup), ]
nrow(babies2)
write.table(babies2,"babies2.txt")
```

```r
babies3=babies[ babies$agegroup==3 & !is.na(babies$agegroup), ]
nrow(babies3)
write.table(babies3,"babies3.txt")
```

```r
babies4=babies[ babies$agegroup==4 & !is.na(babies$agegroup), ]
nrow(babies4)
write.table(babies4,"babies4.txt")
```

```r
babies5=babies[ babies$agegroup==5 & !is.na(babies$agegroup), ]
nrow(babies5)
write.table(babies5,"babies5.txt")
```

```r
babies6=babies[ babies$agegroup==6 & !is.na(babies$agegroup), ]
nrow(babies6)
write.table(babies6,"babies6.txt")
```

```r
par(mai=c(0.5,0.5,0.5,0.5),mfrow=c(3,2)) #by rows
plot(babies1[,5],babies1[,6],main=paste("AGE",label[1]),xlab="height",ylab="weight")
plot(babies2[,5],babies2[,6],main=paste("AGE",label[2]),xlab="height",ylab="weight")
plot(babies3[,5],babies3[,6],main=paste("AGE",label[3]),xlab="height",ylab="weight")
plot(babies4[,5],babies4[,6],main=paste("AGE",label[4]),xlab="height",ylab="weight")
plot(babies5[,5],babies5[,6],main=paste("AGE",label[5]),xlab="height",ylab="weight")
plot(babies6[,5],babies6[,6],main=paste("AGE",label[6]),xlab="height",ylab="weight")
```
