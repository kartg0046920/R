```r
# brain $ FSIQ --> brain [ , 2 ] --> brain[,i]
cat("high std\ttop std")
label = colnames (brain)
for (i in 2:4)
{ 
bm=brain[brain [ , i ]>mean(brain [ , i ]),]
(m=mean(bm [ , i]))
cat(label[i],":",m," ")
bh=brain[brain [ , i ]>quantile(brain [ , i ],0.75),]
(h=mean(bh[ , i ]))
cat(h,"\n")
}
```
