```r
wdbc=read.table("wdbc.txt",header=T,sep=",")
attach(wdbc)
summary(wdbc)
pdf("wdbc.pdf")
sink("sink.txt")
par(mfrow=c(3,4))
label=colnames(wdbc)
for(i in 3:32)
{
hist(wdbc[,i],main=label[i])
st=shapiro.test(wdbc[,i])
cat(label[i],st$p.value,ifelse(st$p.value<0.05,"*",""),"\n")
}
dev.off()
sink()
```
