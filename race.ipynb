```r
set.seed(123)
age=sample(20:60,100, replace=TRUE) 
sex=sample(c("M","F"),100,replace=TRUE)
race=sample(c("WHITE","YELLOW","BLACK"),100,replace=TRUE)
data=data.frame(age,sex,race)
barplot(table(race),col= 4:6)
table(race)
```

```r
barplot(table(sex,race), col =c("pink","blue"), axes= FALSE, beside=TRUE)
legend(0.5,40,c("Female","Male"),col=c("pink","blue"),pch=15)
title(main="Race and Sex Bar Chart",sub="in1988")
axis(2,las=2)
```

```r
boxplot(age~race, main="Age by Race", col="yellow",boxwex=0.3)
boxplot(age[sex== "M"]~race[sex=="M"],col=3,boxwex=0.1,at= c(1.3,2.3,3.3), add=TRUE, axes=F)

boxplot(age[sex== "F"]~race[sex=="F"],col="pink",boxwex=0.1,at= c(0.7,1.7,2.7), add=TRUE, axes=F)
```
